```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import scipy.stats
```

```python
filename = '../../data/donnees_pollution_mortalite_mensuelles.csv'
temp_df = pd.read_csv(filename, sep=";")
polluants = ['no', 'no2', 'o3', 'pm2.5', 'pm10']
nom_lignes = []
data = []
temp = []
for i in range(len(temp_df)):
        nom_lignes.append(temp_df.iloc[i,0])

for j in range(len(polluants) + 1):
    temp = []
    for i in range(len(temp_df)):
        temp.append(temp_df.iloc[i,j+1])
    data.append(temp)
ar = numpy.array(data)
df = pd.DataFrame(ar.transpose(), index = nom_lignes, columns = ['nb_deces'] + polluants)
df
```
